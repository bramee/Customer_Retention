```python
sns.set_theme()
sns.countplot(df['1Gender of respondent'])
```

```python
col = ['22 Ease of navigation in website','23 Loading and processing speed','24 User friendly Interface of the website','25 Convenient Payment methods','26 Trust that the online retail store will fulfill its part of the transaction at the stipulated time','27 Empathy (readiness to assist with queries) towards the customers',
      '32 Shopping online is convenient and flexible','33 Return and replacement policy of the e-tailer is important for purchase decision','39 Offering a wide variety of listed product in several category','47 Getting value for money spent']
plt.figure(figsize = (25,25))
for i in range(1,len(col)+1):
    plt.subplot(5,3,i)
    sns.countplot(x= df[col[i-1]])
plt.tight_layout()
plt.show()
```

* Customers are happy with the easy payment methods in site and website is user- friendly for the customer to do shopping.
* Return and replacement in online shopping is important and customers are stromgly agree that online shopping does that.
* Customers queries are getting resolved through customer care services in e-comm shopping.
* E-comm shopping is nothing but the online shopping and it should be easy and flexible to users and it is and we can conclude that e-comm shopping is easy and flexible for customers.
* Also Customers agreed that it offers a wide variety of products.
* So Customers are satisfied for the money spent on online shop.
